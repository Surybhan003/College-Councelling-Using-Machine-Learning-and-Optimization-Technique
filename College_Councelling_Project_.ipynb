```python
def sort_preferred_branches(df, preferred_branches):
    for preferred_branch, boost in preferred_branches.items():
        df.loc[df['Academic Program Name'].str.lower().str.contains(preferred_branch.lower()), 'Boosted Rank'] = df['Closing Rank'] * (1 - boost)
    df['Boosted Rank'].fillna(df['Closing Rank'], inplace=True)
    sorted_df = df.sort_values(by='Boosted Rank', ascending=True)
    return sorted_df
```

```python
def sort_preferred_branches(df, preferred_branches_iit, preferred_branches_non_iit):
    """
    Sorts the DataFrame by boosted ranks, applying different branch preferences for IITs and non-IITs.
    """
    # Calculate boosted ranks for sorting
    for index, row in df.iterrows():
        branch_boosts = preferred_branches_iit if row['main/advanced'] == 'JEE advance' else preferred_branches_non_iit

        # Apply branch-specific boost if branch is in preference list
        for preferred_branch, boost in branch_boosts.items():
            if preferred_branch.lower() in row['Academic Program Name'].lower():
                df.loc[index, 'Boosted Rank'] = row['Adjusted Rank'] * (1 - boost)
                break  # Stop after the first matching preference

    # Fill non-boosted rows with Closing Rank for consistency
    df['Boosted Rank'].fillna(df['Adjusted Rank'], inplace=True)
    sorted_df = df.sort_values(by='Boosted Rank', ascending=True).reset_index(drop=True)
    return sorted_df
```
